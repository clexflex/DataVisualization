```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Excel file
file_path = 'firemarket.xlsx'
sheet_names = ['Table1', 'Table2', 'Table3', 'Table4', 'Table5', 'Table6', 'Table7']

data = {sheet: pd.read_excel('firemarket.xlsx', sheet_name=sheet) for sheet in sheet_names}
```

```python

df_table1 = data['Table1']

# Plot Market Size and Growth Y-O-Y
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(df_table1['Year'], df_table1['Size in (USD Million)'], 'g-')
ax2.plot(df_table1['Year'], df_table1['Market Size and Y-O-Y'], 'b-')

ax1.set_xlabel('Year')
ax1.set_ylabel('Size in (USD Million)', color='g')
ax2.set_ylabel('Market Size growth Y-O-Y', color='b')

plt.title('Global Electric Fireplace Market Size and Y-O-Y Growth')
plt.show()
```
