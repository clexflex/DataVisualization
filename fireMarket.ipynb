```python
import pandas as pd

# Load the Excel file
file_path = 'firemarket.xlsx'
xls = pd.ExcelFile(file_path)

# Read each table into a DataFrame
table1 = pd.read_excel(xls, sheet_name='Table1')
table2 = pd.read_excel(xls, sheet_name='Table2')
table3 = pd.read_excel(xls, sheet_name='Table3')
table4 = pd.read_excel(xls, sheet_name='Table4')
table5 = pd.read_excel(xls, sheet_name='Table5')
table6 = pd.read_excel(xls, sheet_name='Table6')
table7 = pd.read_excel(xls, sheet_name='Table7')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Set the style
sns.set(style='whitegrid')

# Function to plot Global Market Size Over Time
def plot_global_market_size(table1):
    plt.figure(figsize=(10, 2))
    plt.plot(table1['Year'], table1['Size in (USD Million)'], marker='o')
    plt.title('Global Electric Fireplace Market Size Over Time')
    plt.xlabel('Year')
    plt.ylabel('Market Size (USD Million)')
    plt.grid(True)
    plt.show()
```

```python

# Function to plot Market Size Growth Y-O-Y
def plot_market_growth(table1):
    plt.figure(figsize=(10, 6))
    plt.bar(table1['Year'][1:], table1['Market Size growth Y-O-Y'][1:], color='skyblue')
    plt.title('Market Size Growth Year-Over-Year')
    plt.xlabel('Year')
    plt.ylabel('Growth Rate (%)')
    plt.grid(True)
    plt.show()
```

```python

# Function to plot Regional Market Size Distribution
def plot_regional_distribution(table2):
    table2_melted = table2.melt(id_vars='Region', var_name='Year', value_name='Market Size (USD Million)')
    plt.figure(figsize=(12, 8))
    sns.barplot(data=table2_melted, x='Year', y='Market Size (USD Million)', hue='Region')
    plt.title('Regional Market Size Distribution')
    plt.xlabel('Year')
    plt.ylabel('Market Size (USD Million)')
    plt.legend(title='Region')
    plt.show()
```

```python

# Function to plot North America Market Size and Growth
def plot_na_market(table3):
    fig, ax1 = plt.subplots(figsize=(10, 6))
    
    ax1.set_title('North America Market Size and Growth')
    ax1.set_xlabel('Year')
    ax1.set_ylabel('Market Size (USD Million)')
    ax1.plot(table3['Year'], table3['Size in (USD Million)'], color='tab:blue', marker='o', label='Market Size')
    ax1.tick_params(axis='y')
    
    ax2 = ax1.twinx()
    ax2.set_ylabel('Growth Rate (%)')
    ax2.bar(table3['Year'][1:], table3['YOY growth'][1:], color='skyblue', alpha=0.6, label='YOY Growth')
    ax2.tick_params(axis='y')
    
    fig.tight_layout()
    plt.show()
```

```python

# Function to plot segmented analysis for North America
def plot_segmented_analysis(table, title):
    table_melted = table.melt(id_vars='Segment', var_name='Year', value_name='Market Size (USD Million)')
    plt.figure(figsize=(12, 8))
    sns.barplot(data=table_melted, x='Year', y='Market Size (USD Million)', hue='Segment')
    plt.title(title)
    plt.xlabel('Year')
    plt.ylabel('Market Size (USD Million)')
    plt.legend(title='Segment')
    plt.show()
```

```python

# Plot all the visualizations
plot_global_market_size(table1)
plot_market_growth(table1)
plot_regional_distribution(table2)
plot_na_market(table3)
plot_segmented_analysis(table4, 'North America Market by Country')
plot_segmented_analysis(table5, 'North America Market by Types')
plot_segmented_analysis(table6, 'North America Market by Installations')
plot_segmented_analysis(table7, 'North America Market by Application')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Table 1 Data
years = ['2023', '2024', '2025', '2026', '2027', '2028', '2029', '2030']
size_usd_million = [2065.91, 2123.20, 2184.67, 2253.13, 2333.55, 2432.57, 2558.45, 2722.61]
growth_yoy = [None, 2.78, 2.90, 3.14, 3.57, 4.25, 5.18, 6.42]

# Table 2 Data (Assumed Values)
regions = ['North America', 'Europe', 'Asia-Pacific', 'MEA & Africa', 'South America']
region_data = {
    '2023': [888.75, 450, 450, 150, 127.16],
    '2024': [914.68, 464, 464, 154, 132.52],
    '2025': [942.22, 478, 478, 158, 138.15],
    '2026': [972.36, 493, 493, 163, 144.19],
    '2027': [1006.98, 510, 510, 169, 151.38],
    '2028': [1048.42, 529, 529, 176, 159.46],
    '2029': [1099.78, 554, 554, 184, 170.23],
    '2030': [1164.82, 586, 586, 194, 191.59]
}

# Create DataFrame for Table 2
df_region = pd.DataFrame(region_data, index=regions)

# Visualization for Table 1
fig, ax1 = plt.subplots(figsize=(12, 6))

color = 'tab:blue'
ax1.set_xlabel('Year')
ax1.set_ylabel('Size in USD Million', color=color)
ax1.plot(years, size_usd_million, color=color, marker='o')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Market Size Growth Y-O-Y (%)', color=color)
ax2.plot(years, growth_yoy, color=color, marker='x')
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
plt.title('Global Electric Fireplace Market Size and Growth (2023-2030)')
plt.show()

# Visualization for Table 2
df_region.T.plot(kind='bar', stacked=True, figsize=(14, 8))
plt.title('Global Electric Fireplace Market Size by Region (2023-2030)')
plt.xlabel('Year')
plt.ylabel('Size in USD Million')
plt.legend(title='Region')
plt.show()
```
