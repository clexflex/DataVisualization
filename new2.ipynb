```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Excel file
file_path = 'firemarket.xlsx'
sheet_names = ['Table1', 'Table2', 'Table3', 'Table4', 'Table5', 'Table6', 'Table7']

data = {sheet: pd.read_excel(file_path, sheet_name=sheet) for sheet in sheet_names}
df_table1 = data['Table1']

# Convert columns to appropriate data types if needed
df_table1['Year'] = pd.to_datetime(df_table1['Year'], format='%Y')

# Plot Market Size and Growth Y-O-Y
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot Size in (USD Million)
color = 'tab:green'
ax1.set_xlabel('Year')
ax1.set_ylabel('Size in (USD Million)', color=color)
ax1.plot(df_table1['Year'], df_table1['Size in (USD Million)'], color=color, marker='o', label='Size in (USD Million)')
ax1.tick_params(axis='y', labelcolor=color)

# Create a second y-axis to plot Market Size and Y-O-Y
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Market Size and Y-O-Y', color=color)
ax2.plot(df_table1['Year'], df_table1['Market Size and Y-O-Y'], color=color, marker='s', linestyle='--', label='Market Size and Y-O-Y')
ax2.tick_params(axis='y', labelcolor=color)

# Adding legends
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9))

# Adding gridlines
ax1.grid(True)
ax2.grid(False)

# Setting the title
plt.title('Global Electric Fireplace Market Size and Y-O-Y Growth')

# Display the plot
plt.show()
```

```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Load the Excel file
file_path = 'firemarket.xlsx'
sheet_names = ['Table1', 'Table2', 'Table3', 'Table4', 'Table5', 'Table6', 'Table7']

data = {sheet: pd.read_excel(file_path, sheet_name=sheet) for sheet in sheet_names}
df_table1 = data['Table1']

# Create subplots with secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Add traces
fig.add_trace(
    go.Scatter(x=df_table1['Year'], y=df_table1['Size in (USD Million)'], name='Size in (USD Million)',
               mode='lines+markers', line=dict(color='green')),
    secondary_y=False,
)

fig.add_trace(
    go.Scatter(x=df_table1['Year'], y=df_table1['Market Size and Y-O-Y'], name='Market Size and Y-O-Y',
               mode='lines+markers', line=dict(color='blue', dash='dash')),
    secondary_y=True,
)

# Add figure title
fig.update_layout(
    title_text="Global Electric Fireplace Market Size and Y-O-Y Growth"
)

# Set x-axis title
fig.update_xaxes(title_text="Year")

# Set y-axes titles
fig.update_yaxes(title_text="Size in (USD Million)", secondary_y=False)
fig.update_yaxes(title_text="Market Size and Y-O-Y", secondary_y=True)

# Set y-axis tick interval
fig.update_yaxes(tick0=0, dtick=50, secondary_y=False)
fig.update_yaxes(tickformat=".2%", secondary_y=True)

# Show the plot
fig.show()
```
